# Vamos analiser a Ação do banco Itau

```python
#importando biblioteca

#biblioteca para trabalhar os dados
import pandas as pd

#biblioteca numpy
import numpy as np

#biblioteca que gera os gráficos.
import matplotlib.pyplot as plt

#acessar dados de forma remota
import pandas_datareader.data as web




#comando para importar base de dados da web.
# cota_ibov = web.DataReader("ticker-código-da-moeda/ação", source="fonte", Start="data-inicio", end="data-termino")

cota_itau = web.DataReader("ITUB4.SA", data_source="yahoo", start="12/01/2021", end="12/01/2022")

#OBS: ao ir no site do yahoo finance note que devemos pesquisar a moeda e acrescentala dentro das (" ")
#OBS: a data inicia com dia/mes/ano



display(cota_itau)

#criando o gráfico da ação Banco fechamento de alções do itau
cota_itau["Adj Close"].plot()

#Para exibir gráfico grande
cota_itau["Adj Close"].plot(figsize=(13,12))

```
